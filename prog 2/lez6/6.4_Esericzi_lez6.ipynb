Trasforma le seguenti richiesti in formule per estrarre infromazioni dal dataframe
- Mostrami le vendite maggiori di 10

- Mostrami i dati del 2018

- Mostrami le vendite maggiori di 13 e l'anno è il 2018

- Mostrami tutto TRANNE i casi in cui le vendite sono maggiori di 13 e l'anno è il 2018

- Mostrami i dati dove le vendite divise per 3 sono maggiori di 3

```python
import pandas as pd

sales = pd.DataFrame(
    data={
        "employee": [
            "Katrina",
            "Guanyu",
            "Jan",
            "Roman",
            "Jacqueline",
            "Paola",
            "Esperanza",
            "Alaina",
            "Egweyn",
        ],
        "sales": [14, 17, 6, 12, 8, 3, 7, 15, 5],
        "year": [2018, 2019, 2020, 2018, 2020, 2019, 2019, 2020, 2020],
    }
)
sales
```

```python
#1.Mostrami le vendite maggiori di 10
sales.query('sales > 10')
```

```python
#2.Mostrami i dati del 2018
sales.query('year==2018')
```

```python
#3.Mostrami le vendite maggiori di 13 e l'anno è il 2018
sales.query('sales>13 & year==2018')
```

```python
#4.Mostrami tutto TRANNE i casi in cui le vendite sono maggiori di 13 e l'anno è il 2018
sales.query('not(sales>13 & year==2018)')
```

```python
#5.Mostrami i dati dove le vendite divise per 3 sono maggiori di 3
sales.query('sales//3 >3')
```

```python
#6.Mostrami i dipendenti i cui nomi sono alfabeticamente dopo la J
sales.query('employee > "J"')
```

# 5. Analisi di un dataset di video giochi
1. Quanti videogiochi sono stati pubblicati
2. Quali sono i generi più popolari? Mostralo con un bar plot
3. Plotta l'evoluzione del numero di giochi pubblicati nel tempo
4. Prova a riprodurre questo plot 




```python
# Import dei pacchetti
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Caricamento del dataset
url = 'https://zenodo.org/record/5898311/files/vgsales.csv'
df = pd.read_csv(url)
df.head()
```

```python
# 1. Quanti videogiochi sono stati pubblicati
print(df.Name.count())
#oppure
df.shape[0]
```

```python
# 2. Quali sono i generi più popolari? Mostralo con un bar plot
sns.countplot(data=df,y='Genre',order=df.Genre.value_counts().index)
plt.show()
```

```python
# 3. Plotta l'evoluzione del numero di giochi pubblicati nel tempo
fig, ax = plt.subplots(1,2,figsize=(10, 4))
game_year = df.Year.value_counts().sort_index()
ax[0].plot(game_year.index,game_year.values)
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Games')
ax[1] = sns.kdeplot(data=df, x='Year')
ax[0].set_title('ll numero di giochi pubblicati nel tempo')
ax[1].set_title('ll numero di giochi pubblicati nel tempo')
plt.tight_layout()
plt.show()
```

```python
# 4. Prova a riprodurre questo plot
df_Genre = df.pivot_table(values=['NA_Sales','EU_Sales','JP_Sales','Other_Sales'],index='Genre',aggfunc='sum')
df_Genre = df_Genre.merge(df.groupby('Genre')['Genre'].agg(popularity = 'size'),on='Genre')
df_Genre.sort_values(by='popularity',ascending=False,inplace=True)
df_Genre.iloc[0:5,0:-1].plot(kind='bar',stacked=True,title='Vendite per regione per genere',ylabel='Milioni di unità')
plt.show()
```
